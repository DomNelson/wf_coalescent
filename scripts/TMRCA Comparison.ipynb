```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
tmrca_ARGON_file = os.path.expanduser('~/project/wf_coalescent/results/ARGON/N200samplesize2rec1e-08mut1e-08length1.0_tmrca.txt')
tmrca_msprime_wf_file = os.path.expanduser('/Users/dnelson/project/wf_coalescent/results/msprime_wf/N200samplesize2rec1e-08mut1e-08length1_tmrca.txt')

msprime_wf_tmrca = np.genfromtxt(tmrca_msprime_wf_file)
ARGON_tmrca = np.genfromtxt(tmrca_ARGON_file)


fudge_factor = 1

bins, edges = np.histogram(msprime_wf_tmrca / fudge_factor, bins=10)
plt.plot(edges[:-1], bins / np.sum(bins), label='msprime_wf')

bins, edges = np.histogram(ARGON_tmrca, bins=10)
plt.plot(edges[:-1], bins / np.sum(bins), label='ARGON')

x = 1
p = 1. / 200
# i = 2
# N = 200
n = np.arange(0, 3000, 1)

# P_Tgtt = (1 - i * (1-i)/(2 * N)) ** (N*t)
plt.plot(n, stats.binom.cdf(x, n, p)/2, 'o', ms=2, label='binom pmf')
# plt.plot(n, (1 - 1 / 200) ** (n), 'o', ms=2, label='e**-t')
# plt.plot(t, P_Tgtt, 'o', ms=2, label='e**-t')

plt.legend()

plt.show()
```
